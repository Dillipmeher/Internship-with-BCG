```python
import matplotlib.pyplot as plt

# Assuming 'churn' is a binary variable with values 'Yes' and 'No'
churn_counts = client_df['churn'].value_counts()

plt.figure(figsize=(8, 6))
bars = churn_counts.plot(kind='barh', color=['#00A6A6', '#F40076'], width=0.8, linewidth=0.3)  # Adjust the linewidth to reduce bar thickness
plt.xlabel("Count", labelpad=14)
plt.ylabel("Target Variable", labelpad=14)
plt.title("Count of TARGET Variable per category", y=1.02)

# Annotate bars with their values and percentages
for bar in bars.patches:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width()} ({bar.get_width() / len(client_df) * 100:.2f}%)',
             va='center', ha='left', color='black')

# Remove background grid lines
bars.axes.get_xaxis().set_visible(False)
bars.grid(False, axis='y')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each channel sales category
channel_sales_counts = client_df['channel_sales'].value_counts()

# Calculate churn rate for each channel sales category
churn_rate_per_channel = client_df.groupby('channel_sales')['churn'].mean()

# Plotting the distribution of channel sales and churn rate
plt.figure(figsize=(12, 6))
ax = channel_sales_counts.plot(kind='bar', color='#00A6A6', width=0.4, linewidth=0.3)  # Adjust the linewidth to reduce bar thickness
ax.set_xlabel("Channel Sales", labelpad=14)
ax.set_ylabel("Count of Customers", labelpad=14)
plt.title("Distribution of Channel Sales with Churn Rate", y=1.02)

# Annotate bars with their values (count of customers)
for bar in ax.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height()}',
             ha='center', va='bottom', color='black')

# Create a twin y-axis for churn rate
ax2 = ax.twinx()
ax2.set_ylabel("Churn Rate", labelpad=14)

# Plotting churn rate as a line plot
churn_rate_per_channel.plot(kind='line', marker='o', color='red', ax=ax2)

# Annotate churn rate points on the line plot
for i, rate in enumerate(churn_rate_per_channel):
    ax2.text(i, rate, f'{rate:.2f}', ha='center', va='bottom', color='black')

plt.xticks(range(len(channel_sales_counts)), channel_sales_counts.index, rotation=45, ha='right')  # Rotate x-labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for better visualization

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
client_df = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\BCGX internship\client_data (1).csv")

# Separate numeric and non-numeric columns
numeric_columns = client_df.select_dtypes(include=['int64', 'float64']).columns
non_numeric_columns = client_df.select_dtypes(exclude=['int64', 'float64']).columns

# Create histograms for numeric columns
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=client_df, x=column, kde=True)
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()

# Create count plots for non-numeric columns
for column in non_numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=client_df, x=column)
    plt.title(f"Count of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```
